```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print("Original dataset shape:", df.shape)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())

print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
print("TotalCharges data type before:", df['TotalCharges'].dtype)
print("Sample values:", df['TotalCharges'].unique()[:10])

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

print("Missing values after conversion:", df['TotalCharges'].isnull().sum())
print("TotalCharges data type after:", df['TotalCharges'].dtype)
```

```python
median_charges = df['TotalCharges'].median()
df['TotalCharges'] = df['TotalCharges'].fillna(median_charges)

print("Missing values after filling:", df['TotalCharges'].isnull().sum())
```

```python
print("Columns before removal:", df.columns.tolist())

df = df.drop('customerID', axis=1)

print("Columns after removal:", df.columns.tolist())
print("Dataset shape after removal:", df.shape)
```

```python
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)

if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates removed. New shape:", df.shape)
else:
    print("No duplicates found.")
```

```python
print("Final data types:")
print(df.dtypes)
```

```python
categorical_columns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'Churn']

print("Unique values in categorical columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()}")
```

```python
df.to_csv('telco_cleaned.csv', index=False)
print("Cleaned dataset saved as 'telco_cleaned.csv'")
```
